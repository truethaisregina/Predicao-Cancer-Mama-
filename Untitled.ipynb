```python
base_filtro = base [['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
var_resposta =  base["diagnosis"]
base = base_filtro
```

```python
def categoriza(cat):
    if cat == "B":
        return 0
    else:
        return 1
    
    base['diagnosis']= base['diagnosis'].apply(categoriza)
```

#### Separando TREINO E TESTE

```python
from sklearn.model_selection import train_test_split
```

```python
X_all = base_filtro
y_all = var_resposta

X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size = 0.20 ,random_state = 79831)

print('Numero de observaÃ§oes do treino:', len(X_train))
print('Numero de observaÃ§oes da teste:',len(X_test))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

#### Tentando fazer a MODELAGEM E VALIDAÃ‡ÃƒO DOS RESULTADOS

```python
clf = DecisionTreeClassifier(random_state=123,max_depth=3,min_samples_leaf=30)
```

```python
modelo = clf.fit(X_train,y_train)
```

```python
scores = cross_val_score(clf, X_train, y_train, cv=5)
scores
```

```python
print(f'Accuracy {scores.mean() :.2f} (+/- {scores.std() * 2 :.2f})')
```
